#### Basic Git Steps

- initialize repository - track chances to files in a certain directory
- add files to "staging area"
- commit milestones
- Edit files, add, commit, repeat

#### Remote Repositories
- "push" changes to a remote location
- "Clone" an entire repository
- "pull" changes from a remote location

```bash
git config --global user.name "name"
git config --global user.email "email"
git config --global color.ui "auto"
```

* `pwd` - present working directory
* `cd` - change directory
* `ls` - list files in curent directory
* `mkdir` - make a directory


```bash
mkdir git_test
cd git_test/
git init
ls -al 
```

The final command shows all objects and information about files, hidden and otherwise.

```bash
nano my_file.txt
```

Enter some text into this file. Close nano and enter the following command:

```bash
git status
```

This gives the output:
```
On branch master

No commits yet

Untracked files:
  (use "git add <file>..." to include in what will be committed)

	my_file.txt

nothing added to commit but untracked files present (use "git add" to track)
```

Create a new file:

```bash
touch my_file2.txt 
```

```bash
git status
```
Output:
```
On branch master

No commits yet

Changes to be committed:
  (use "git rm --cached <file>..." to unstage)

	new file:   my_file.txt

Untracked files:
  (use "git add <file>..." to include in what will be committed)

	my_file2.txt
	my_file3.txt
```

Add more files:
```bash
git add *.txt 

git diff
```

Output:
```
diff --git a/my_file.txt b/my_file.txt
index df20303..24b7100 100644
--- a/my_file.txt
+++ b/my_file.txt
@@ -2,3 +2,4 @@ test
 
 more stuff
 
+even more stuff
```

```bash
git log
```
